```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt, exp
from scipy.integrate import odeint
import scipy.special as sc
%matplotlib inline
```

```python
#Fonctions de conversion utiles
def s2day(t) :
    '''Conversion de t en seconde vers t en jours'''
    return t/(3600*24)

def s2year(t) :
    '''Conversion de t en seconde vers t en années'''
    return t/(3600*24*365)


```

```python
#Constantes physiques
c = 3.e+8           #Vitesse de la lumière 
h = 6.626e-34       #Planck
kb = 1.380e-23      #Boltzmann
Jy = 10.e-26        #1 jansky en SI
sigma = 5.67e-8     #constante de Stefan
a = 4*sigma/c
D_MPc = 3.086e22    #Conversion 1Mpc en m

#Constantes pour le soleil
M_sun = 1.989e+30
R_sun = 696340e+3
L_sun = 3.827e+26


#Paramètre pour le terme de chauffage
A = 6.6e+42/M_sun   
alpha = 1.3
```

```python
#Constantes du problème
M_ej = 1.e-2*M_sun  #Masse éjectée
v_exp = 0.15*c      #Vitesse d'expension 
R_0 = 1.e+4         #Rayon initial
kappa = 0.5         #Opacité
D = 40*D_MPc

#Constantes du model
k1 = 1.             #Ajustement du model pour temps radiatif 
```

```python
#Calcul des temps et longueurs caractéristiques
t_star = sqrt(3*M_ej*kappa/(4*pi*c*v_exp))
R_neb = sqrt(3*M_ej*kappa/(4*pi))
t_neb = R_neb/v_exp

print("Constantes caractéristiques du probleme :")
print("t* = " + str("%.2e"%t_star) + " s = " + str("%.2f"%s2day(t_star)) + " jours")
print("t_neb = " + str("%.2e"%t_neb) + " s = " + str("%.2f"%s2day(t_neb)) + " jours")
print("R_neb = " + str("%.2e"%R_neb) + " m = " + str("%.2f"%(R_neb/R_sun)) + " Rayons solaire")
```

```python
#Dérivée temporelle de l'énergie interne que l'on utilise dans le modèle suivant
def E_dot(t) :
    return A*M_ej*pow(t,-alpha)


#Equation différentielle à résoudre avec odeint
def model(L,t) :
    dLdt = (E_dot(t)-L)*k1*t/pow(t_star,2)
    return dLdt


#Integration de l'équation différentielle pour L(t) sur t dans [0, 1.3*t_neb]
t_list = np.linspace(1.e-6, 1.3*int(t_neb), num=10000)
L0 = 0
L = odeint(model, L0, t_list)
```

```python
## Quantitées dépendantes explicitement du temps

#Rayon de la supernova qui est une fonction affine du temps 
def R(t) :
    return v_exp*t 

#Libre parcours moyen des photons en fonction du temps
def l_bar(t) :
    rho = 3*M_ej/(4*pi*pow(R(t),3))
    return 1/(rho*kappa)



## Quantitées calculées à partir de la luminosité

#T_list est la liste de la température interne sur un maillage temporel t_list
def T_list(t_list) :
    res = []
    for i in range(len(t_list)) :
        t = t_list[i]
        L_t = L[i]
        T_t = pow(L_t/(4*pi*R(t)*sigma*l_bar(t)),1/4)
        res+= [T_t]
    return res
        
#Ts_list est la liste de la température de surface sur un maillage temporel t_list
def Ts_list(t_list) :
    res=[]
    T = T_list(t_list)
    for i in range(len(t_list)) :
        t = t_list[i]
        T_t = T[i]
        Ts_t = pow(l_bar(t)/R(t),1/4)*T_t
        res+= [Ts_t]
    return res

#P_list est la liste de la pression sur un maillage temporel t_list
def P_rad_list(t_list) :
    T = T_list(t_list)
    return [a*pow(t,4)/3 for t in T]

def P_gaz_list(t_list) :
    P_gaz_list = []
    T = T_list(t_list)
    for i in range(len(t_list)) :
        t = t_list[i]
        Temp = T[i]
        rho = 3*M_ej/(4*pi*pow(R(t),3))
        P_gaz = rho*kb*Temp/0.5   #le 0.5 vient du mu*mp de l'Hydrogène
        P_gaz_list += [P_gaz]
    return P_gaz_list

#U_list est la liste de l'énergie interne sur un maillage temporel t_list
def U_list(t_list) : 
    T = T_list(t_list)
    res = []
    for i in range(len(T)) :
        t = t_list[i]
        Temp = T[i]
        res += [4*pi*pow(R(t),3)*a*pow(Temp,4)/3]
    return res

#Expression de l'intensité spécifique (dans un modèle de corps noir, à sa surface)
def B_list(nu,Ts_arg_list):
    res = []
    for i in range(len(Ts_arg_list)):
        Ts_arg = Ts_arg_list[i]
        Bnu = (2*h*pow(nu,3)/pow(c,2)) / (np.exp((h*nu)/(kb*Ts_arg)) - 1)
        res += [Bnu]
    return res
```

```python
# Calculs des grandeurs de la kilonova

T = T_list(t_list)
Ts = Ts_list(t_list)
P_rad = P_rad_list(t_list)
P_gaz = P_gaz_list(t_list)
U = U_list(t_list)
```

```python
## Représentation grphique de L bolométrique ainsi que L pour différentes longueurs d'ondes

Nlim = 7550 #Utile pour tracer en pointillés après t_neb

plt.figure(figsize=[14,8])
plt.yscale('log')
plt.ylim(1.e5,5.e7)
plt.xlabel('temps (days)')
plt.ylabel('$L/L_{\odot }$')

#Plot de E point
plt.plot(s2day(t_list), E_dot(t_list)/L_sun, color='orange', label='$\dot{E}$')
#Ajout des lignes pointillées pour t_star et t_neb
plt.axvline(x=s2day(t_star),color='gray',linestyle='--')
plt.axvline(x=s2day(t_neb),color='gray',linestyle='--')
#Ajout legende pour t_star et t_neb
plt.text(int(s2day(t_star))+0.2,4.e7,'$t_{*}$',fontsize=15)
plt.text(int(s2day(t_neb)) -0.5,4.e7,'$t_{neb}$',fontsize=15)

#Plot Luminosité bolométrique
plt.plot(s2day(t_list[:Nlim]),L[:Nlim]/L_sun,color='black', label='$Bolometric$')
plt.plot(s2day(t_list[Nlim:]),L[Nlim:]/L_sun,linestyle='--',color='black')

#Plot Luminosité Spécifiques
lambda_Lspecifique = [1000.e-9, 650.e-9, 400.e-9]
color_Lspecifique = ['#FF00DC', '#FF2D00', '#3E00FF']
label_Lspecifique = ['1$\mu$m', '650nm', '400nm']
for i in range (len(lambda_Lspecifique)) :
    lambd = lambda_Lspecifique[i]
    nu = c/lambd
    Bnu = np.array(B_list(nu,Ts[1:]))
    L_specifique = np.array([4*pow(pi,2)*pow(R(t_list[k+1]),2)*Bnu[k] for k in range(len(t_list)-1)])
    plt.plot(s2day(t_list[:Nlim]),nu*L_specifique[:Nlim]/L_sun,color=color_Lspecifique[i], label = label_Lspecifique[i])
    
plt.legend(loc="upper right")
```

La figure ci-dessus permet de se rendre compte de plusieurs choses. 

- On voit que entre $t_{star}$ et $t_{neb}$, l'approximation $L = \dot{E}$ est bonne

```python
## Température et énergie interne

plt.figure(figsize=[15,8])

#Plot de la température 
plt.subplot(121)
plt.xlabel("temps (jours)")
plt.ylabel("Température (K)")
plt.yscale('log')
plt.plot(s2day(t_list[1:]),T[1:],label='T(t)', color ='#FF3105')
plt.plot(s2day(t_list[1:]),Ts[1:],label='Ts(t)', color='#FFB305')
plt.legend()

#Plot de l'énergie interne
plt.subplot(122)
plt.xlabel("temps (jours)")
plt.ylabel("Energie interne (Joules)")
plt.yscale('log')
plt.plot(s2day(t_list[1:]),U[1:],label='U(t)',color='#BF05FF')
plt.legend()
```

```python
## Comparaison de la pression de radiation et de la pression interne

plt.figure(figsize=[15,8])

#Plot des Pression de radiation et 
plt.xlabel("temps (jours)")
plt.ylabel("Pression (Pa)")
plt.yscale('log')
plt.plot(s2day(t_list[1:]),P_rad[1:],label='P_rayonnement(t)',color='#086A20')
plt.plot(s2day(t_list[1:]),P_gaz[1:],label='P_gaz(t)',color='#58CB74')
plt.legend()
```

```python
## Comparaison du libre parcours moyen et du rayon

plt.figure(figsize=[15,8])

#Plot du rayon et du libre parcours moyen 
plt.xlabel("temps (jours)")
plt.ylabel("Rayon (m)")
plt.yscale('log')
plt.plot(s2day(t_list[1:]),[R(t) for t in t_list[1:]],label='R(t)',color='#08366A')
plt.plot(s2day(t_list[1:]),[l_bar(t) for t in t_list[1:]],label='libre parcours moyen',color='#0AD5C6')
plt.text(int(s2day(t_neb)) -0.5,4.e7,'$t_{neb}$',fontsize=15)

plt.axvline(x=s2day(t_neb),color='gray',linestyle='--')
plt.legend()
```

## Décomposition de l'intensité selon la fréquence

Sous l'hypothèse que la kilonova se comporte comme un corps noir, on peut facilement exploiter l'expression de la température de surface $T_s = (\frac{\bar l}{R})^{\frac{1}{4}}T$ pour tracer l'intensité spécifique pour chaque longueur d'onde à l'aide de la loi de Planck. Le but ici est d'obtenir un graphe similaire aux données de la figure 1 de l'article sur la kilonova GW170817.

```python
## Tracé de l'intensité spécifique pour différentes longueurs d'onde

#Longueurs d'ondes et légendes correspondantes
lambd = [2190e-9,1630e-9,1020e-9,900e-9,806e-9,658e-9,464e-9,365e-9]
lettres = ['K','H','Y','Z','i','r','g','u']
colors = [ [8/255,8/255,8/255], [165/255,0/255,38/255], [244/255,109/255,67/255], [253/255,174/255,97/255], [171/255,217/255,233/255], [116/255,173/255,209/255], [69/255,117/255,180/255], [49/255,54/255,149/255] ]


plt.figure(figsize=[10,8])

for i in range(len(lambd)):

    #Intensité spécifique
    nu = c/lambd[i]
    Bnu = B_list(nu,Ts)

    #Conversion en jansky et en magnitude AB
    magAB = [-2.5*np.log( pi*Bnu[i]*pow((R(t_list[i])/D),2)/(3631*Jy) )/np.log(10) for i in range(len(Bnu))]

    plt.plot(s2day(t_list[1:]),magAB[1:],label=lettres[i], color=colors[i])

plt.title("Magnitudes pour chaque longueur d'onde")
plt.xlabel('temps (days)')
plt.ylabel("Mag AB")
plt.ylim(20,30)
plt.gca().invert_yaxis()
plt.legend()
```

On obtient un graphe dont l'allure est dans un premier temps comparable à celle de la figure 1, les petites longueurs d'ondes sont étouffées en premier comme l'on pourrait s'y attendre dans un modèle de refroidissement. Notre modèle simple nous a donc permis de capter l'essentiel de la dynamique. Cependant, l'article se penche en particulier sur la dynamique des intensités aux grandes longueurs d'ondes, là où les UV subissent une décroissance assez immédiate, les infrarouges semblent être interrompus dans leur décroissance par une période de plateau.

La partie suivante consiste à essayer d'expliquer ce phénomène par un modèle à 2 éjectas présenté dans l'article.

## Raffinement du modèle par une hypothèse à 2 éjectas

On considère que la kilonova est séparée en 2 éjectas: un constitué de noyaux légers ($A \lessapprox 140$) que l'on nommera "bleu", et l'autre constitué de de noyaux lourds ($A \gtrapprox 140$) que l'on nommera "rouge". L'intensité qui nous parvient serait la somme des intensités pour chacun des éjectas.

Pour chaque éjecta on prendra les données proposées par l'article dans le modèle à 2 éjectas.



```python
## Résolution par un modèle à 2 éjectas

# Calcul des paramètres (tirés du modèle à 2 composantes de l'article) et détermination de t_néb = max(t_néb_rouge,t_néb_bleue)

# Bleu
M_ej_b = 0.014*M_sun  #Masse éjectée
v_exp_b = 0.266*c     #Vitesse d'expension
kappa_b = 0.05        #Opacité

t_star_b = sqrt(3*M_ej*kappa_b/(4*pi*c*v_exp_b))
R_neb_b = sqrt(3*M_ej*kappa_b/(4*pi))
t_neb_b = R_neb_b/v_exp_b

# Rouge
M_ej_r = 0.036*M_sun  #Masse éjectée
v_exp_r = 0.123*c     #Vitesse d'expension
kappa_r = 0.335       #Opacité

t_star_r = sqrt(3*M_ej*kappa_r/(4*pi*c*v_exp_r))
R_neb_r = sqrt(3*M_ej*kappa_r/(4*pi))
t_neb_r = R_neb_r/v_exp_r

# t_neb
t_neb = max(t_neb_b, t_neb_r)





# Paramètres d'une kilonova "bleue"
M_ej, v_exp, kappa, t_star, R_neb = M_ej_b, v_exp_b, kappa_b, t_star_b, R_neb_b

# Calcul des grandeurs de la kilonova "bleue"
t_list = np.linspace(1.e-6, 1.3*int(t_neb), num=10000)

L = odeint(model, L0, t_list)

L_b = odeint(model, L0, t_list)
T_b = T_list(t_list)
Ts_b = Ts_list(t_list)





# Paramètres d'une kilonova "rouge" (tirées du modèle à 2 composantes de l'article)
M_ej, v_exp, kappa, t_star, R_neb = M_ej_r, v_exp_r, kappa_r, t_star_r, R_neb_r

# Calcul des grandeurs de la kilonova "rouge"
t_list = np.linspace(1.e-6, 1.3*int(t_neb), num=10000)

L = odeint(model, L0, t_list)

L_r = odeint(model, L0, t_list)
T_r = T_list(t_list)
Ts_r = Ts_list(t_list)





# Luminosité totale:
L_tot = [L_b[i] + L_r[i] for i in range(len(t_list))]





# Courbes de luminosité

plt.figure(figsize=[14,8])
plt.yscale('log')
plt.xlabel('temps (days)')
plt.ylabel('$L/L_{\odot }$')

L_b_sun = [L_b[i]/L_sun for i in range(len(t_list))]
L_r_sun = [L_r[i]/L_sun for i in range(len(t_list))]
L_tot_sun = [L_tot[i]/L_sun for i in range(len(t_list))]

plt.plot(s2day(t_list), L_b_sun, color='blue', label='KN bleue')
plt.plot(s2day(t_list), L_r_sun, color='red', label='KN rouge')
plt.plot(s2day(t_list), L_tot_sun, color='black', label='Modèle à 2 éjectas')
    
plt.legend(loc="upper right")
```

Un rapide tracé de la luminosité nous permet de constater que l'éjecta "bleu" contribue d'avantage durant la phase initiale, et vice-versa pour l'éjecta "rouge". Pour avoir une meilleure compréhension du phénomène, on trace par la suite la décomposition en intensité spécifique dans un modèle de corps noir.

```python
## Tracé de l'intensité spécifique pour différentes longueurs d'onde dans les modèles bleus et rouges

#Longueurs d'ondes et légendes correspondantes
lambd = [2190e-9,1630e-9,1020e-9,900e-9,806e-9,658e-9,464e-9,365e-9]
lettres = ['K','H','Y','Z','i','r','g','u']
colors = [ [8/255,8/255,8/255], [165/255,0/255,38/255], [244/255,109/255,67/255], [253/255,174/255,97/255], [171/255,217/255,233/255], [116/255,173/255,209/255], [69/255,117/255,180/255], [49/255,54/255,149/255] ]

plt.figure(figsize=[24,7])

for i in range(len(lambd)):

    #Intensités spécifiques
    nu = c/lambd[i]
    Bnu_b = B_list(nu,Ts_b)
    Bnu_r = B_list(nu,Ts_r)

    #Conversion en jansky et en magnitude AB
    magAB_b = [-2.5*np.log( pi*Bnu_b[i]*pow((R(t_list[i])/D),2)/(3631*Jy) )/np.log(10) for i in range(len(Bnu_b))]
    magAB_r = [-2.5*np.log( pi*Bnu_r[i]*pow((R(t_list[i])/D),2)/(3631*Jy) )/np.log(10) for i in range(len(Bnu_r))]
    magAB_tot = [-2.5*np.log( pi*(Bnu_b[i]+Bnu_r[i])*pow((R(t_list[i])/D),2)/(3631*Jy) )/np.log(10) for i in range(len(Bnu_r))]

    plt.subplot(1,3,1)
    plt.plot(s2day(t_list[1:]),magAB_b[1:],label=lettres[i], color=colors[i])
    
    plt.subplot(1,3,2)
    plt.plot(s2day(t_list[1:]),magAB_r[1:],label=lettres[i], color=colors[i])
    
    plt.subplot(1,3,3)
    plt.plot(s2day(t_list[1:]),magAB_tot[1:],label=lettres[i], color=colors[i])

plt.subplot(1,3,1)
plt.title("Magnitude de l'éjecta bleu")
plt.xlabel('temps (days)')
plt.ylabel("Mag AB")
plt.ylim(20,26)
plt.gca().invert_yaxis()
plt.legend()

plt.subplot(1,3,2)
plt.title("Magnitude de l'éjecta rouge")
plt.xlabel('temps (days)')
plt.ylabel("Mag AB")
plt.ylim(20,26)
plt.gca().invert_yaxis()
plt.legend()

plt.subplot(1,3,3)
plt.title("Magnitude de la kilonova à 2 éjectas")
plt.xlabel('temps (days)')
plt.ylabel("Mag AB")
plt.ylim(20,26)
plt.gca().invert_yaxis()
plt.legend()
```

Sur les courbes ci-dessus on arrive à une conclusion similaire que dans l'article: un modèle à 1 éjecta, que ce soit bleu ou rouge, ne permet pas d'expliquer l'entièreté de la dynamique de l'évolution de l'intensité. Il faut bien le modèle plus complet pour pouvoir expliquer la brève stagnation des grandes longueurs d'ondes.

On remarque que cette stagnation n'est pas aussi flagrante pour les bandes K et H que dans l'article, c'est possiblement dû à la simplicté de notre modèle, qui se retrouve moins adapté pour les valeurs de masse et de vitesse d'éjecta proposées par l'article. L'éjecta "bleu" étant responsable de la montée initiale, augmenter la valeur de sa masse dans le code ci-dessus permet de faire apparaitre la dynamique en question sur les bandes K et H.

L'article conclut sur l'idée que si le modèle à 2 éjectas permet une meilleure explication de la courbe de lumière, alors l'éjecta "rouge" constitué de noyaux lourds est indispensable, mettant en avant l'idée que c'est dans les kilonovae que ces éléments sont créés.
